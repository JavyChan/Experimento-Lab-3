```python
velocidad_esfera_1 = Altura/tiempo_medio1 #cm/s
velocidad_esfera_2 = Altura/tiempo_medio2 #cm/s
velocidad_esfera_3 = Altura/tiempo_medio3 #cm/s

print(velocidad_esfera_1,velocidad_esfera_2,velocidad_esfera_3)
```

```python
density_1 = 49.7/(4/3*np.pi) #cgs
density_2 = 16.3/(1.72)
density_3 = 8.4/(1.14)
density_array = np.array([density_1,density_2,density_3])

diameter_1 = 2.0 #cm
diameter_2 = 1.5 
diameter_3 = 1.3
diameter_array = np.array([diameter_1,diameter_2,diameter_3])

mean_velocity_array = np.array([velocidad_esfera_1,velocidad_esfera_2,velocidad_esfera_3])

mean_time_array = np.array([tiempo_medio1,tiempo_medio2,tiempo_medio3])

density_slime = 0.80 #cgs

mu_array = g_cgs*diameter_array**2*(density_array-density_slime)/(18*mean_velocity_array)
```

```python
print(mu_array,density_array)

std_array = np.array([std1,std2,std3])

mu_err = mu_array/(mean_time_array)*std_array
print(mu_err)
```

```python
speed_slime = 274/(np.pi*(1.12)**2*296)

l = 2.25

Reynolds = density_slime*speed_slime*l/(mu_array)

print(Reynolds)
```

```python

a1,a0 = np.polyfit(Reynolds,mu_array,1)

l1,l0 = np.polyfit((Reynolds),np.log10(mu_array),1)

print(l1,l0)

plt.loglog(Reynolds,a1*Reynolds+a0,label="Least-squares (linear)",color="red")
plt.plot(Reynolds,10**(l1*Reynolds+l0),label = "Least-squares (logarithmic)")

plt.errorbar(Reynolds,mu_array, yerr = mu_err,label =  "Puntos Viscosidad vs Re", fmt = "o")

plt.xlabel("Número de Reynolds del Slime")
plt.ylabel("Viscosidad")
plt.title("Gráfico log-log Reynolds vs Viscosidad")

plt.legend()
plt.savefig("EXPERIMENTO_grafico-loglog.pdf")
```
